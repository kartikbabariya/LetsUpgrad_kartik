# Question 14

```python
sns.barplot(x='matchType',y='killPoints',data=df)
plt.xticks(rotation=90);
```

From above plot we concluded that if you play in match type "normal-duo-fpp" then the chances of getting kill points more. Also,if you play in "squad-fpp" then then you also get better amount of points but less than normal-duo-fpp mode. Except normal-duo-fpp and squad-fpp you get less amount of point in other match type mode. In "flaretpp,crashtpp & flarefpp" you can less amout of kill poiuts as compare to others. So if you want to get more kill points the i suggest you to play "normal-duo-fpp" match type mode.
